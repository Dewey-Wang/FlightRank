```python
from typing import List, Optional
import polars as pl
import numpy as np

def compute_ranked_average_selected_from_submissions(
    submission_paths: List[str],
    mode: str = "average",  # "average" or "weighted"
    weights: Optional[List[float]] = None
) -> pl.DataFrame:
    """
    多個 submission.parquet 做 selected 加權平均或普通平均，並依 ranker_id 排名。

    參數:
    - submission_paths: List of parquet 檔案路徑
    - mode: "average" | "weighted"
    - weights: list of float, 與 submission_paths 對應的權重（mode 為 weighted 時必填）

    回傳:
    - Polars DataFrame：Id, ranker_id, __index_level_0__, selected
    """
    if not submission_paths:
        raise ValueError("請至少提供一個 submission 檔案路徑")

    if mode not in ["average", "weighted"]:
        raise ValueError("mode 僅能是 'average' 或 'weighted'")

    if mode == "weighted":
        if weights is None:
            raise ValueError("加權模式需要提供 weights")
        if len(weights) != len(submission_paths):
            raise ValueError("weights 的長度需等於 submission_paths 數量")
        weights = np.array(weights) / np.sum(weights)

    dfs = []
    for i, path in enumerate(submission_paths):
        print(f"✅ 讀取: {path}")
        df = pl.read_parquet(path)
        df = df.select([
            "Id",
            "ranker_id",
            "__index_level_0__",
            pl.col("selected").alias(f"selected_{i}")
        ])
        dfs.append(df)

    merged_df = dfs[0]
    for df in dfs[1:]:
        merged_df = merged_df.join(df, on=["Id", "ranker_id", "__index_level_0__"], how="inner")

    selected_cols = [f"selected_{i}" for i in range(len(dfs))]

    if mode == "average":
        merged_df = merged_df.with_columns([
            pl.mean_horizontal(selected_cols).alias("selected")
        ])
    else:  # weighted
        weighted_exprs = [pl.col(col) * w for col, w in zip(selected_cols, weights)]
        merged_df = merged_df.with_columns([
            sum(weighted_exprs).alias("selected")
        ])

    merged_df = merged_df.with_columns(
        pl.col("selected")
        .rank(method="ordinal", descending=True)
        .over("ranker_id")
        .alias("selected")
    )

    merged_df = merged_df.sort("Id")
    print(f"✅ 完成 {len(merged_df)} 筆 {mode} + 排名 計算")
    return merged_df.select(["Id", "ranker_id", "selected", "__index_level_0__"])

# 使用方式：
# avg_df = compute_ranked_average_selected_from_submissions(submission_paths, mode="average")
# weighted_df = compute_ranked_average_selected_from_submissions(submission_paths, mode="weighted", weights=[0.1, 0.2, ..., 0.1])
```
